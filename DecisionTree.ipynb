```python
import pandas as pd
```

```python
datas=pd.read_csv("50_Startups.csv")
```

```python
datas=pd.get_dummies(datas,drop_first=True)
```

```python
datas.columns
```

```python
independent=datas[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
```

```python
dependent=datas[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
regr=DecisionTreeRegressor(criterion='mae',max_features='log2',splitter='random')
regr=regr.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as pl
from sklearn import tree
tree.plot_tree(regr)
pl.show()
```

```python
y_pred=regr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```
